```python
import pandas as pd
import numpy as np
```

```python
from google.colab import files

uploaded = files.upload()

df = pd.read_csv(next(iter(uploaded)))
```

```python
print("First 5 Rows")
print(df.head())

print("\nDataset Information")
print(df.info())

print("\nSummary Statistics")
print(df.describe())

print("\nMissing Values")
print(df.isnull().sum())
```

```python
# Fill numeric columns with median
numeric_cols = df.select_dtypes(include=np.number).columns

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())

# Fill categorical columns with mode
categorical_cols = df.select_dtypes(include='object').columns

for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

print(df.isnull().sum())
```

Q1. What is the average, median, and standard deviation of house prices?

```python
print("Average Price :", df["Price"].mean())

print("Median Price :", df["Price"].median())

print("Standard Deviation :", df["Price"].std())
```



---



Q2. Which number of bedrooms is most common?

```python
print("Most Common Bedrooms")

print(df["number of bedrooms"].mode())
```



---



Q3. Check if the price data is skewed.

```python
skew = df["Price"].skew()

print("Skewness =", skew)

if skew > 0:
    print("Price is Positively Skewed")
elif skew < 0:
    print("Price is Negatively Skewed")
else:
    print("Price is Symmetric")
```



---



Q4. What is the average price for each number of bedrooms?

```python
avg_price = df.groupby("number of bedrooms")["Price"].mean()

print(avg_price)
```



---



Q5.  What is the relationship between living area and price? (Use : Area without Basement).

```python
correlation = df["Area of the house(excluding basement)"].corr(df["Price"])

print("Correlation =", correlation)
```



---



Q6. Identify any anomalies where houses have high prices but low areas.

```python
high_price = df["Price"].quantile(0.90)

low_area = df["Area of the house(excluding basement)"].quantile(0.25)

anomalies = df[
    (df["Price"] > high_price) &
    (df["Area of the house(excluding basement)"] < low_area)
]

print(anomalies)
```



---



Q7. Compare average price based on number of floors and houses with or without waterfront.

```python
result = df.groupby(
    ["number of floors", "waterfront present"]
)["Price"].mean()

print(result)
```

Q8. Identify the minimum and maximum house price. What does this indicate?



```python
print("Minimum Price")

print(df["Price"].min())

print("\nMaximum Price")

print(df["Price"].max())
```

Q9. Which location (zipcode/area) has the highest average price?



```python
postal = df.groupby("Postal Code")["Price"].mean()

print(postal)

print("\nPostal Code with Highest Average Price")

print(postal.idxmax())

print("Average Price")

print(postal.max())
```
